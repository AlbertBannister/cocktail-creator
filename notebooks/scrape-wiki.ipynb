```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
resp = requests.get("https://en.wikipedia.org/wiki/List_of_cocktails")
soup = BeautifulSoup(resp.text)
soup
```

```python
content = soup.find(id="mw-content-text")
links = [x.a for x in content.find_all("li") if x.a]
links
```

```python
links[6].get("href")
```

```python
paths = [a.get("href") for a in links if a.get("href")]
WIKI_URL = "https://en.wikipedia.org"
urls = [WIKI_URL + p for p in paths]
urls
```

```python
cocktail_url = urls[0]
page = requests.get(cocktail_url).text
soup = BeautifulSoup(page)
soup
```

```python
# We find the cocktail infobox table and extract the name and recipe attributes
infoboxes = soup.find_all("table",class_ ="infobox")
recipes=[]
for infobox in infoboxes:
    headers = infobox.find_all("th", class_="infobox-label")
    data = infobox.find_all("td", class_="infobox-data")

    title = infobox.find(class_="infobox-title")
    print(title.text)
    # for h, d in zip(headers, data):
        # print("attr: {}, val: {}".format(h.text, [x.text for x in d.ul if len(x.text) > 2] if d.ul else d.text))
    
    recipe = {h.text: [x.text for x in d.ul if len(x.text) > 2] if d.ul else d.text for h, d in zip(headers, data)}

    recipe["name"] = title.text
    recipes.append(recipe)
```

```python
recipes = []
for url in urls:
    page = requests.get(url).text
    soup = BeautifulSoup(page)
    for infobox in soup.find_all("table", class_="infobox"):
        try:
            headers = infobox.find_all("th", class_="infobox-label")
            data = infobox.find_all("td", class_="infobox-data")

            title = infobox.find(class_="infobox-title")
            print(title.text)
            # for h, d in zip(headers, data):
                # print("attr: {}, val: {}".format(h.text, [x.text for x in d.ul if len(x.text) > 2] if d.ul else d.text))
            
            recipe = {h.text: [x.text for x in d.ul if len(x.text) > 2] if d.ul else d.text for h, d in zip(headers, data)}

            recipe["name"] = title.text
            recipes.append(recipe)
        except Exception as e:
            print(url)
            print(e)
            

    
```
